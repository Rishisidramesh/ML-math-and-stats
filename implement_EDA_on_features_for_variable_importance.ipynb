```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_excel("preprocessed_titanic_dataset.xlsx")
df.isnull().sum()
```

```python
df['Fare'].value_counts()
```

```python
mean = np.mean(df['Fare'])
median = np.median(df['Fare'])
mode = stats.mode(df['Fare'])
```

```python
mean, median,mode
```

```python
df['Fare'].plot(kind='hist',bins = 20)
plt.xlabel("fare prize")
plt.ylabel('frequency')
plt.title("fare distribution")
plt.show()
```

```python
df.groupby('Survived')['Fare'].mean()
```

```python
df.groupby('Survived')['Fare'].median()
```

```python
df.groupby("Pclass")["Fare"].mean()
```

```python
stats.skew(df['Fare'])
```

```python
df['Fare_log'] = np.log1p(df['Fare'])
stats.skew(df['Fare_log']) 
```

```python
df["Fare_log"].plot(kind='hist',bins=20)
plt.title("log_fare distribution")
plt.xlabel("log(fare)")
plt.ylabel("frequncy")
plt.show()
```

```python
df.groupby("Survived")["Fare_log"].mean()
```

```python
df.groupby("Pclass")["Fare_log"].mean()
```
